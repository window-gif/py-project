```python
import pandas as pd
import jieba
import nltk
from pyecharts import options as opts
from pyecharts.charts import *
from pyecharts.globals import SymbolType
from pyecharts import options as opts
import matplotlib.pyplot as plt
%matplotlib
def word_croud(df):
    cut_words=jieba.lcut(''.join(df['标题'].values),cut_all=False)
    path = r'C:\Users\22442\Desktop\停用词表.txt'
    fd = open(path,encoding = 'utf-8')
    stop_word = fd.read() 
    new_data = []
    for word in cut_words:
        if word not in stop_word:
            new_data.append(word)
    freqlist = nltk.FreqDist(new_data)
    freqlist = freqlist.most_common()
    words = freqlist
    c = WordCloud()
    c.add("", words, word_size_range=[20, 80])
    c.set_global_opts(title_opts=opts.TitleOpts(title="美团标题标签可视化"))
    return c

def score_medium(df):
    data2 = df.groupby('分类')['评分'].agg(['mean','median'])
    x = list(data2.index)
    y = list(data2.iloc[:,1])
    chart = Bar()  
    chart.add_xaxis(x)  
    chart.add_yaxis("评分", y)  
    chart.set_global_opts(
        title_opts = opts.TitleOpts(
        title = '不同项目中中位数评分情况',
        subtitle = '中位数'),
        xaxis_opts=opts.AxisOpts(
            axislabel_opts=opts.LabelOpts(rotate=45)
        )
    )
    return chart

def score_average(df):
    data2 = df.groupby('分类')['评分'].agg(['mean','median'])
    x = list(data2.index)
    y = list(data2.iloc[:,0].round(2))
    chart = Line()  
    chart.add_xaxis(x)  
    chart.add_yaxis("评分", y)  
    chart.set_global_opts(
        title_opts = opts.TitleOpts(
        title = '不同项目中平均数评分情况',
        subtitle = '平均数'),
        xaxis_opts=opts.AxisOpts(
            axislabel_opts=opts.LabelOpts(rotate=45)
        )
    )
    return chart

def sales_sum(df):
    data2 = df.groupby('分类')['销量'].agg(['sum'])
    x = list(data2.index)
    y = list(data2.iloc[:,0])
    chart = Bar()  
    chart.add_xaxis(x)  
    chart.add_yaxis("销量", y)  
    chart.set_global_opts(
        title_opts = opts.TitleOpts(
        title = '不同项目中销量和评分情况',
        subtitle = '销量和'),
        xaxis_opts=opts.AxisOpts(
            axislabel_opts=opts.LabelOpts(rotate=45)
        )
    )
    return chart

def price_pie(df):
    prince_bin = pd.cut(df['价格'],bins = [0,10,30,50,70,100,500,3000],
                    include_lowest = True,right = False,
                    labels = ['<10','10-30','30-50','50-70','70-100','100-500','>500'])
    df_prince = pd.value_counts(prince_bin)
    data_prince = [list(i) for i in zip(df_prince.index,list(df_prince[:]))]
    pie = Pie()
    pie.add('',data_prince)
    pie.set_global_opts(title_opts=opts.TitleOpts(title="Pie-商品价格比例"))
    pie.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c}"))
    return pie

path = r'C:\Users\22442\Desktop\美团.xlsx'
df = pd.read_excel(path)
df = df.drop_duplicates(inplace = False)
df.reset_index(drop = True,inplace = True)
df = df.fillna(0)
line_df = df['销量'].astype(str)
line_df = line_df.str.replace('已售','')
line_df = line_df.astype(int)
df['销量'] = pd.to_numeric(line_df)
line_df = df['价格'].str.replace('元','')
line_df = line_df.astype(float)
df['价格'] = pd.to_numeric(line_df)
grid1 = (
    Grid()
    .add(score_medium(df), grid_opts=opts.GridOpts(pos_left='5%', pos_right='10%', pos_top='5%', pos_bottom='55%'))
)

grid2 = (
    Grid()
    .add(score_average(df), grid_opts=opts.GridOpts(pos_left='55%', pos_right='5%', pos_top='5%', pos_bottom='55%'))
)

grid3 = (
    Grid()
    .add(sales_sum(df), grid_opts=opts.GridOpts(pos_left='8%', pos_right='20%', pos_top='15%', pos_bottom='20%'))
)

grid4 = (
    Grid()
    .add(price_pie(df), grid_opts=opts.GridOpts(pos_left='60%', pos_right='5%', pos_top='15%', pos_bottom='20%'))
)

page = Page(layout=Page.SimplePageLayout)  # 使用 SimplePageLayout 布局，使得图表在同一行显示
page.add(grid1,grid2, grid3, grid4)  # 将图表1、2、3、4添加到Page中
page.render("meituan.html", notebook=False)
```
